# Apply Transformation

```python
def reverse_name(name):
  return name[::-1]
```

```python
reverse_name('Ravi')
```

```python
reverse_name('Sarthak')
```

```python
name= 'Data Science'
reverse_name(name)
```

```python
data['name']=data['name'].apply(reverse_name)
```
