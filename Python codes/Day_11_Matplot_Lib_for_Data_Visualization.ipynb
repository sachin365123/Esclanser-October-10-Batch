# Grid Line chart

```python
np.arange(0.0,2.0,0.1) # starting index and ending index and skipping value
```

```python
t=np.arange(0.0,2.0,0.01)
s=1+np.sin(2*np.pi*t)
plt.plot(t,s)

plt.xlabel('time (second) ---> ')
plt.ylabel('velocity (m/sec) ---> ')
plt.title('Physics theory graph --->')
plt.grid(True)
plt.savefig('Day10_LineChart.png') # save a plot and check the directory
plt.show()
```

# Bar chart

```python
y= [3, 10, 7, 5, 3, 16, 8]
x= range(len(y))
width = 1/4
plt.bar(x,y,width, color='green')
plt.show()
```

# Bubble Chart

```python
labels = 'Data Scinece', 'Machine Learning', 'Maths' , 'Excel'
sizes = [15,25,45,15 ]
# sum(sizes)
explod = (0.1,0,0,0.3)
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode= explod, labels=labels, autopct = '%1.2f%%',shadow=False, startangle=-250)
ax1.axis('equal')
plt.show()
```

# Scatter Chart

```python
N = 50
# Generate random numbers
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=0.7)
plt.show()
```
