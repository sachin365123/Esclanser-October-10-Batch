```python
import numpy as np
```

```python
# Define a sharpening kernel
kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 2]])

sharpened_img = cv2.filter2D(img, -1, kernel)
```

```python
cv2_imshow(sharpened_img)
```
